```
from PIL import Image
import matplotlib.pyplot as plt
import os

img_path="room.jpg"

if os.path.exists(img_path):
    img=Image.open(img_path)
    plt.imshow(img)
    plt.title("Input Image")
    plt.axis("off")
    plt.show()
else:
    print("Place your input image in the notebook folder as room.jpg")
```

```
from PIL import Image
import matplotlib.pyplot as plt
import os

candidates=[
    "outputs/room.png",
    "outputs/depth_room.png"
]

found=False
for p in candidates:
    if os.path.exists(p):
        depth=Image.open(p)
        plt.imshow(depth)
        plt.title("Predicted Depth Map")
        plt.axis("off")
        plt.show()
        found=True
        break

if not found:
    print("No depth map found. Update the filename if necessary.")
```

## Observations

- Depth Anything predicts a **depth map**, not a full 3D model.
- COLMAP reconstructs geometry using classical computer vision.
- NeRF represents a scene as a neural radiance field.
- Gaussian Splatting renders scenes efficiently using optimized Gaussian primitives.
